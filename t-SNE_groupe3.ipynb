```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Chargement du dataset de chiffres manuscrits

digits = load_digits()

# Instanciation d'un objet TSNE avec une perplexité de 30 et une valeur de learning rate 200

tsne = TSNE (n_components=2, perplexity=30, learning_rate=200)

# Réduction de la dimensionnalité des données avec le t-SNE 

digits_tsne = tsne.fit_transform(digits.data)

# Création d'un graphe scatter plot des valeurs transformées

plt.scatter (digits_tsne[:, 0], digits_tsne[:, 1], c=digits.target) 
plt.show()
```

```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import pandas as pd

# Chargement du dataset de chiffres manuscrits

digits = load_digits()

# Instanciation d'un objet TSNE avec une perplexité de 30 et une valeur de learning rate 200

tsne = TSNE (n_components=2, perplexity=30, learning_rate=200)

# Réduction de la dimensionnalité des données avec le t-SNE

digits_tsne = tsne.fit_transform(digits.data)

# Création d'un DataFrame avec les données transformées et les étiquettes

df = pd.DataFrame(digits_tsne, columns=['x', 'y'])
df['label'] = digits.target

# Création d'un graphe à barres des valeurs transformées

df.plot.bar(x='x', y='y', color='red')
plt.show()
```
